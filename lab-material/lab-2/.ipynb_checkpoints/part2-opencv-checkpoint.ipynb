## Section 2: Basic image interfacing with openCV, channels, flipping, resizing, shape drawing

```python
import cv2
img = cv2.imread("tree.png")
plt.imshow(img)
```

```python
print(type(img))
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
resized_img = cv2.resize(img, (1500,1000))
plt.imshow(resized_img)
```

```python
flip_img = cv2.flip(img, 1)
plt.imshow(flip_img)
```

```python
cv2.rectangle(resized_img, pt1=(550, 100), pt2=(1300,750), color=(255,0,0), thickness=10)
plt.imshow(resized_img)
```

## Section 3: Filtering in Images, low-pass, Sobel

```python
bw_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(bw_img, cmap="gray")
```

```python
highpass_filter = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])
filtered_img = cv2.filter2D(bw_img, -1, highpass_filter)
plt.imshow(filtered_img, cmap="gray")
```

```python
lowpass_filter = np.ones((15,15))*1/225
filtered_img = cv2.filter2D(bw_img, -1, lowpass_filter)
plt.imshow(filtered_img, cmap="gray")
```
